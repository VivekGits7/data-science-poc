# Pandas - Data Analysis & Manipulation

**What:** The core Python library for working with structured/tabular data (like Excel or SQL tables).

**When to use:**
- Reading data from CSV, JSON, Excel, SQL databases
- Cleaning, filtering, sorting, and transforming data
- Grouping, aggregating, and pivoting data
- Merging/joining multiple datasets
- Preparing data before feeding it to ML models

**Key concept:** Everything revolves around the `DataFrame` (a table with rows and columns) and `Series` (a single column).

```python
# 1. Setup
import pandas as pd
import numpy as np

print(f"Pandas version: {pd.__version__}")
```

```python
# 2. Create DataFrame
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "age": [25, 30, 35, 28, 32],
    "department": ["Engineering", "Sales", "Engineering", "HR", "Sales"],
    "salary": [75000, 55000, 82000, 60000, 58000],
    "joined": ["2022-01-15", "2021-06-20", "2020-03-10", "2023-02-01", "2021-11-05"],
})

print(df)
```

```python
# 3. Read Data from Files
# df = pd.read_csv("data.csv")
# df = pd.read_json("data.json")
# df = pd.read_excel("data.xlsx")
# df = pd.read_sql("SELECT * FROM table", connection)

print("Uncomment the line matching your file format")
```

```python
# 4. Basic Info
print("=== Shape ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n=== Data Types ===")
print(df.dtypes)

print("\n=== Info ===")
df.info()
```

```python
# 5. head() and tail()
print("=== First 3 rows ===")
print(df.head(3))

print("\n=== Last 2 rows ===")
print(df.tail(2))
```

```python
# 6. describe() - Statistical Summary
print(df.describe())
```

```python
# 7. Select Columns
print("=== Single column (Series) ===")
print(df["name"])

print("\n=== Multiple columns (DataFrame) ===")
print(df[["name", "salary"]])
```

```python
# 8. loc[] and iloc[]
print("=== loc (by label) ===")
print(df.loc[0])  # First row
print("\n", df.loc[1:3, ["name", "salary"]])  # Rows 1-3, specific columns

print("\n=== iloc (by position) ===")
print(df.iloc[0])  # First row
print("\n", df.iloc[1:4, [0, 3]])  # Rows 1-3, columns 0 and 3
```

```python
# 9. Filtering Rows
print("=== Salary > 60000 ===")
print(df[df["salary"] > 60000])

print("\n=== Engineering department ===")
print(df[df["department"] == "Engineering"])

print("\n=== Multiple conditions (AND) ===")
print(df[(df["salary"] > 55000) & (df["age"] < 35)])

print("\n=== Multiple conditions (OR) ===")
print(df[(df["department"] == "Sales") | (df["department"] == "HR")])
```

```python
# 10. String Filtering
print("=== Name contains 'a' (case insensitive) ===")
print(df[df["name"].str.contains("a", case=False)])

print("\n=== Name starts with 'A' ===")
print(df[df["name"].str.startswith("A")])
```

```python
# 11. Sorting
print("=== Sort by salary (ascending) ===")
print(df.sort_values("salary"))

print("\n=== Sort by salary (descending) ===")
print(df.sort_values("salary", ascending=False))

print("\n=== Sort by multiple columns ===")
print(df.sort_values(["department", "salary"], ascending=[True, False]))
```

```python
# 12. Add / Modify Columns
df["bonus"] = df["salary"] * 0.1
df["tax"] = df["salary"] * 0.2
df["net_salary"] = df["salary"] + df["bonus"] - df["tax"]

print(df[["name", "salary", "bonus", "tax", "net_salary"]])
```

```python
# 13. apply() - Custom Function
df["seniority"] = df["age"].apply(lambda x: "Senior" if x >= 30 else "Junior")

print(df[["name", "age", "seniority"]])
```

```python
# 14. groupby() - Aggregate Data
print("=== Average salary by department ===")
print(df.groupby("department")["salary"].mean())

print("\n=== Multiple aggregations ===")
print(df.groupby("department").agg({
    "salary": ["mean", "min", "max"],
    "age": "mean",
    "name": "count",
}))
```

```python
# 15. value_counts() / unique() / nunique()
print("=== Value counts ===")
print(df["department"].value_counts())

print(f"\nUnique departments: {df['department'].unique()}")
print(f"Number of unique: {df['department'].nunique()}")
```

```python
# 16. Handle Null Values
df_with_nulls = df.copy()
df_with_nulls.loc[0, "salary"] = None
df_with_nulls.loc[2, "department"] = None

print("=== Check nulls ===")
print(df_with_nulls.isnull().sum())

print("\n=== Fill nulls ===")
print(df_with_nulls["salary"].fillna(df_with_nulls["salary"].mean()))

print("\n=== Drop rows with nulls ===")
print(df_with_nulls.dropna())
```

```python
# 17. Rename and Drop Columns
print("=== Rename ===")
df_renamed = df.rename(columns={"name": "full_name", "salary": "annual_salary"})
print(df_renamed.columns.tolist())

print("\n=== Drop columns ===")
df_dropped = df.drop(columns=["bonus", "tax", "net_salary"])
print(df_dropped.columns.tolist())
```

```python
# 18. Merge / Join DataFrames
dept_df = pd.DataFrame({
    "department": ["Engineering", "Sales", "HR", "Marketing"],
    "budget": [500000, 300000, 200000, 250000],
})

merged = pd.merge(df, dept_df, on="department", how="left")
print(merged[["name", "department", "salary", "budget"]])
```

```python
# 19. Concatenate DataFrames
new_employees = pd.DataFrame({
    "name": ["Frank", "Grace"],
    "age": [27, 33],
    "department": ["Marketing", "Engineering"],
    "salary": [52000, 78000],
})

combined = pd.concat([df[["name", "age", "department", "salary"]], new_employees], ignore_index=True)
print(combined)
```

```python
# 20. Loop Through Rows
for index, row in df.head(3).iterrows():
    print(f"Row {index}: {row['name']} earns {row['salary']}")
```

```python
# 21. Export Data
df.to_csv("output.csv", index=False)
df.to_json("output.json", orient="records", indent=2)
# df.to_excel("output.xlsx", index=False)  # needs openpyxl

print("Exported to output.csv and output.json")
```
