# Matplotlib - Data Visualization

**What:** The core Python plotting library. Creates static, animated, and interactive charts — line, bar, scatter, pie, histogram, box plot, and more.

**When to use:**
- Visualizing trends, patterns, and distributions in data
- Creating charts for reports, presentations, and dashboards
- Comparing datasets visually (grouped bars, multi-line charts)
- Plotting ML model results (loss curves, confusion matrices)
- Saving publication-quality charts as PNG, PDF, SVG

**Key concept:** Everything uses `plt` (pyplot) — call `plt.plot()`, `plt.bar()`, `plt.show()`, etc.

```python
# 1. Setup
import matplotlib.pyplot as plt
import numpy as np

print("Matplotlib ready!")
```

```python
# 2. Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

plt.figure(figsize=(8, 4))
plt.plot(x, y, color="blue", marker="o", linestyle="-", linewidth=2, label="Sales")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales (in thousands)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Multiple Lines on Same Chart
months = ["Jan", "Feb", "Mar", "Apr", "May"]
product_a = [10, 15, 13, 18, 20]
product_b = [8, 12, 17, 14, 22]

plt.figure(figsize=(8, 4))
plt.plot(months, product_a, marker="o", label="Product A", color="#2196F3")
plt.plot(months, product_b, marker="s", label="Product B", color="#FF5722")
plt.title("Product Comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 4. Bar Chart
categories = ["Python", "JavaScript", "Java", "C++", "Go"]
popularity = [35, 28, 10, 12, 5]
colors = ["#4CAF50", "#FFC107", "#F44336", "#2196F3", "#9C27B0"]

plt.figure(figsize=(8, 4))
plt.bar(categories, popularity, color=colors, edgecolor="black", linewidth=0.5)
plt.title("Programming Language Popularity")
plt.xlabel("Language")
plt.ylabel("Popularity (%)")
plt.show()
```

```python
# 5. Horizontal Bar Chart
categories = ["Python", "JavaScript", "Java", "C++", "Go"]
popularity = [35, 28, 20, 12, 5]

plt.figure(figsize=(8, 4))
plt.barh(categories, popularity, color="#2196F3")
plt.title("Language Popularity (Horizontal)")
plt.xlabel("Popularity (%)")
plt.show()
```

```python
# 6. Grouped Bar Chart
categories = ["Q1", "Q2", "Q3", "Q4"]
revenue = [100, 120, 130, 150]
expenses = [80, 90, 100, 110]

x = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(8, 4))
plt.bar(x - width/2, revenue, width, label="Revenue", color="#4CAF50")
plt.bar(x + width/2, expenses, width, label="Expenses", color="#F44336")
plt.xticks(x, categories)
plt.title("Revenue vs Expenses")
plt.legend()
plt.show()
```

```python
# 7. Scatter Plot
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(8, 5))
plt.scatter(x, y, c=y, cmap="viridis", alpha=0.7, edgecolors="black", linewidth=0.5)
plt.colorbar(label="Y Value")
plt.title("Scatter Plot with Color Map")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 8. Histogram
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color="#2196F3", edgecolor="black", alpha=0.7)
plt.title("Distribution of Test Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.axvline(data.mean(), color="red", linestyle="--", label=f"Mean: {data.mean():.1f}")
plt.legend()
plt.show()
```

```python
# 9. Pie Chart
labels = ["Desktop", "Mobile", "Tablet", "Other"]
sizes = [45, 35, 15, 5]
explode = (0.05, 0.05, 0, 0)
colors = ["#2196F3", "#4CAF50", "#FFC107", "#9E9E9E"]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, explode=explode, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.title("Website Traffic by Device")
plt.show()
```

```python
# 10. Subplots - Multiple Charts in One Figure
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Dashboard", fontsize=16)

# Line chart
axes[0, 0].plot([1, 2, 3, 4, 5], [10, 20, 15, 25, 30], marker="o")
axes[0, 0].set_title("Line Chart")
axes[0, 0].grid(True, alpha=0.3)

# Bar chart
axes[0, 1].bar(["A", "B", "C", "D"], [25, 40, 30, 55], color="#4CAF50")
axes[0, 1].set_title("Bar Chart")

# Scatter plot
x = np.random.randn(50)
y = np.random.randn(50)
axes[1, 0].scatter(x, y, color="#FF5722", alpha=0.6)
axes[1, 0].set_title("Scatter Plot")

# Histogram
axes[1, 1].hist(np.random.normal(0, 1, 500), bins=20, color="#9C27B0", alpha=0.7)
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
# 11. Area Chart (Stacked)
months = np.arange(1, 13)
product_a = [20, 25, 30, 35, 40, 45, 42, 38, 35, 40, 45, 50]
product_b = [15, 18, 20, 22, 25, 28, 30, 27, 25, 28, 30, 35]

plt.figure(figsize=(10, 5))
plt.fill_between(months, product_a, alpha=0.5, label="Product A", color="#2196F3")
plt.fill_between(months, product_b, alpha=0.5, label="Product B", color="#4CAF50")
plt.title("Revenue Over Time")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 12. Box Plot
np.random.seed(42)
data = [np.random.normal(50, 10, 100), np.random.normal(60, 15, 100), np.random.normal(45, 8, 100)]

plt.figure(figsize=(8, 4))
bp = plt.boxplot(data, labels=["Group A", "Group B", "Group C"], patch_artist=True)
colors = ["#2196F3", "#4CAF50", "#FFC107"]
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color)
plt.title("Score Distribution by Group")
plt.ylabel("Score")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 13. Available Styles
print("Available styles:")
for style in plt.style.available:
    print(f"  - {style}")
```

```python
# 14. Using a Style
plt.style.use("seaborn-v0_8-darkgrid")

x = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.title("Styled Chart")
plt.legend()
plt.show()

# Reset to default
plt.style.use("default")
```

```python
# 15. Save Chart to File
plt.figure(figsize=(8, 4))
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], marker="o", color="#2196F3")
plt.title("Chart to Save")
plt.grid(True, alpha=0.3)

plt.savefig("my_chart.png", dpi=150, bbox_inches="tight")
plt.savefig("my_chart.pdf", bbox_inches="tight")
print("Saved as my_chart.png and my_chart.pdf")
plt.show()
```
